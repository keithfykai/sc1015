```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
sb.set() 
```

```python
traindata = pd.read_csv('train.csv')
traindata.head()
```

```python
pdata = pd.DataFrame(traindata[["LotArea","GrLivArea", "TotalBsmtSF", "GarageArea", "SalePrice"]])
pdata.describe()
```

```python
Q1 = pdata.quantile(0.25)
Q3 = pdata.quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
((pdata < lower_fence) | (pdata > upper_fence)).sum()
```

#### Lot Area has the highest number of outliers

```python
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = pdata["LotArea"], orient = "h")
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = pdata["GrLivArea"], orient = "h")
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = pdata["TotalBsmtSF"], orient = "h")
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = pdata["GarageArea"], orient = "h")
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = pdata["SalePrice"], orient = "h")
```

```python
f = plt.figure(figsize=(16, 8))
sb.histplot(data = pdata["LotArea"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = pdata["GrLivArea"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = pdata["TotalBsmtSF"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = pdata["GarageArea"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = pdata["SalePrice"])
```

#### Lot Area is skewed to the right (positive skew)

```python
pdata.corr()
```

#### 'GrLivArea' and 'GarageArea' (as they have the highest correlation (closest to 1) with 'SalePrice')

```python
garagedata = pd.DataFrame(traindata["GarageCond"])
print("Data type : ", type(garagedata))
print("Data dims : ", garagedata.size)
garagedata.head()
```

```python
print(traindata['GarageType'].value_counts())
sb.catplot(y = 'GarageType', data = traindata, kind = "count")
```

```python
YearRemodAdd = pd.DataFrame(traindata["YearRemodAdd"])
print("Data type : ", type(YearRemodAdd))
print("Data dims : ", YearRemodAdd.size)
```

```python
SalePrice = pd.DataFrame(traindata["SalePrice"])
print("Data type : ", type(SalePrice))
print("Data dims : ", SalePrice.size)
```

```python
jointDF = pd.concat([YearRemodAdd, SalePrice], axis = 1).reindex(SalePrice.index)
jointDF
```

```python
sb.jointplot(data = jointDF, x = "SalePrice", y = "YearRemodAdd", height = 12)
```

#### Generally, newly remodelled garages have higher sale prices
