### Extracting the 5 variables and plotting their boxplots

```python
p1data = house_prices[['LotArea','GrLivArea','TotalBsmtSF','GarageArea']]
p1data
```

```python
p1data.describe()
```

```python
q1 = p1data.quantile(0.25)
q3 = p1data.quantile(0.75)
iqr = q3-q1
((p1data < q1 - 1.5*iqr) | (p1data > q3 + 1.5*iqr)).sum()
```

Therefore the variable with the most outliers is LotArea with 69 outliers.

```python
f = plt.figure(figsize=(20,5))
sb.boxplot(data=p1data, orient='v').set(ylim=(0,20000))
```

```python
f = plt.figure(figsize=(16, 8))
sb.histplot(data = p1data["LotArea"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = p1data["GrLivArea"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = p1data["TotalBsmtSF"])
f = plt.figure(figsize=(16, 8))
sb.histplot(data = p1data["GarageArea"])
```

```python
p1data.skew(axis=0,skipna=True)
```

From the data, LotArea is the most skewed from a regular normal distribution. With a high positive skew of 12.207688.

```python
jointplot = pd.concat([house_prices[['SalePrice']],p1data],axis=1)
jointplot
```

```python
jointplot.corr()
```

```python
sb.heatmap(jointplot.corr(),vmin=-1,vmax=1,annot=True,fmt=".2f")
```

## Problem 2: Analysis of Categorical Variables

In this problem, your job is to analyze the following categorical variables in the data and their relationship with SalePrice.

['MSSubClass', 'Neighborhood', 'BldgType', 'OverallQual']

a) Which of these variables has the highest number of levels? Which of the levels has the highest number of houses?

b) Choose the top two variables that you think will help us the most in predicting ‘SalePrice’ of houses in this data. (Plot the boxplot)

### Hints and Pointers:

Levels for a categorical variable means the number of unique values. For example, gender has 3 levels, F, M, O.

Each level of a categorical variable may contain a number of datapoints. For example, 14 M found in a dataset.

When you want to find relationship between a numeric variable and a categorical one, you can’t do Correlation.

Check box-plot function in seaborn carefully – there is a way to plot a numeric vs a categorical in the two axes.

Think: If there was a relationship between Salary and Gender, what would box-plot of Salary be across F, M, O?

### Extracting the 4 Variables:

```python
p2data = house_prices[['MSSubClass','Neighborhood','BldgType','OverallQual']]
```

```python
print(p2data[["MSSubClass"]].nunique())
print(p2data[['Neighborhood']].nunique())
print(p2data[['BldgType']].nunique())
print(p2data[['OverallQual']].nunique())
```

There are 15 levels in MSSubClass.

There are 25 levels in Neighborhood.

There are 5 levels in BldgType.

There are 10 levels in OverallQual.

```python
p2data[['Neighborhood']].value_counts()
```

### Solution to Problem 2:

a) Neighborhood has the highest number of levels with 25. NAmes has the highest number of houses: 225.

```python
f = plt.figure(figsize=(20,10))
order = house_prices.groupby('MSSubClass')['SalePrice'].median().sort_values().index
sb.boxplot(data=house_prices, orient='v', x='MSSubClass',y='SalePrice',order=order)

NeighborhoodBP = plt.figure(figsize=(20,10))
order = house_prices.groupby('Neighborhood')['SalePrice'].median().sort_values().index
sb.boxplot(data=house_prices, orient='v', x='Neighborhood',y='SalePrice',order=order)

f = plt.figure(figsize=(20,5))
order = house_prices.groupby('BldgType')['SalePrice'].median().sort_values().index
sb.boxplot(data=house_prices, orient='v', x='BldgType',y='SalePrice',order=order)

f = plt.figure(figsize=(20,10))
order = house_prices.groupby('OverallQual')['SalePrice'].median().sort_values().index
sb.boxplot(data=house_prices, orient='v', x='OverallQual',y='SalePrice',order=order)
```

Note that sorting by median is same as sorting by the levels, denoting a strong increasing trend of SalePrice with increase in OverallQual.

b) The top 2 variables that would help us the most in predicting 'SalePrice' of houses in this data would be Neighborhood and OverallQual as there is lesser overlapping of box plots on the vertical axis. Additionally the medians of these 2 plots are not overlapping as much as BldgType.

## Problem 3: Interesting Questions for EDA (Exploratory Data Analysis)

a) Does the SalePrice of a house get affected by whether it has a Garage or not? Justify your answer using EDA.

For Garage variables, there are a few to choose from, the most obvious and important one being whether it has a garage or not. 

Garage variables: 
    
'GarageType': Garage location, 
    
'GarageYrBlt': Year garage was built, 

'GarageFinish': Interior finish of the garage

'GarageCars': Size of garage in car capacity

'GarageArea': Size of garage in square feet

'GarageQual': Garage quality
		
'GarageCond': Garage condition


For houses with no garage, "GarageType" and "GarageYearBlt" would be NULL.

### Hints and Pointers

Check the data_description file very carefully and pick the variables you want to work with for these problems.

In some cases, it is alright to create new variables out of the original ones, especially if they help you analyze.

Do keep an eye out for missing values in the variables that you tackle and see what you can make out of those.

```python
p3data = house_prices[['SalePrice','GarageType','GarageYrBlt','GarageFinish','GarageCars','GarageArea','GarageQual','GarageCond']].copy()
p3data
```

```python
p3data
```

```python
p3data['GarageExist'] = np.where(p3data[['GarageQual']].isnull(), "No", "Yes")
p3data
```

```python
GarageType = pd.DataFrame(house_prices['GarageType'])
GarageType.value_counts()
```

```python
p3data.isna().sum()
```

Therefore from the data, there are 81 NULL values or 81 houses without Garages.

```python
f = plt.figure(figsize=(20,10))
sb.boxplot(data=p3data, orient='v',x='GarageExist',y='SalePrice')
```

3a) As seen in the boxplot above, we can tell that when the garage exists, the saleprice will be higher as compared to having no garage. Hence we can conclude that the SalePrice of the house will get affected whether it has a Garage or not. Additionally, the mean and median of the SalePrice for houses with Garages is higher than those without Garages.
